```python
#Codsoft 
#Task 1
# --> TITANIC SURVIVAL PREDICTION  <--
#Visualizing data
import pandas as p
import matplotlib.pyplot as mp
import seaborn as s
```

```python
# Load the dataset from local system
titanicdf = p.read_csv(r"C:\Users\jagad\Downloads\Titanic-Dataset.csv")
```

```python
# Display summary statistics
print("Summary Statistics:")
#print(titanicdf.describe())
titanicdf.describe()
```

```python
# Display missing values
print("\nMissing Values:")
#print(titanicdf.isnull().sum())
titanicdf.isnull().sum()
```

```python
# Visualize survival count
s.countplot(x='Survived', data=titanicdf)
mp.title("Survival Count")
mp.show()
```

```python
# Visualize survival count by sex
s.countplot(x='Survived', hue='Sex', data=titanicdf)
mp.title("Survival Count by Sex")
mp.show()
```

```python
# Visualize survival count by passenger class
s.countplot(x='Survived', hue='Pclass', data=titanicdf)
mp.title("Survival Count by Passenger Class")
mp.show()
```

```python
# Visualize age distribution
s.histplot(titanicdf['Age'], bins=20, kde=True)
mp.title("Age Distribution")
mp.show()
```
